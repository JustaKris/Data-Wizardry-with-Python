# Pandas Fundamentals

Learn essential pandas operations for data manipulation.

```python
import pandas as pd
import numpy as np
```

## Reading Data

Learn to load data from various sources.

```python
# Example: Read CSV
# df = pd.read_csv('data.csv')

# Create sample data instead
df = pd.DataFrame({
    'Product': ['A', 'B', 'C', 'A', 'B'],
    'Sales': [100, 150, 200, 120, 180],
    'Region': ['East', 'West', 'East', 'West', 'East']
})
df
```

## Data Selection and Filtering

```python
# Select columns
df[['Product', 'Sales']]
```

```python
# Filter rows
df[df['Sales'] > 150]
```

## GroupBy Operations

```python
# Group and aggregate
df.groupby('Product')['Sales'].agg(['sum', 'mean', 'count'])
```
